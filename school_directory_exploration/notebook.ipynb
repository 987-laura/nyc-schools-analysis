# Clean column names

```python
df.columns = (
    df.columns
      .str.lower()
      .str.replace(" ", "_")
      .str.replace(r"[^a-z0-9_]", "", regex=True)
)
```

# Group and summarize
- Count of schools per borough (again unique schools)
- Average number of students per borough
- Summary of grade_span_max grouped by borough (use the method that provides you with basic statistics about that column)

## Schools per borough

```python
df_schools_per_borough = df.groupby("borough")["dbn"].nunique()
df_schools_per_borough
```

## Average number of students per borough 

```python
df_avg_students_per_borough = df.groupby("borough")["total_students"].mean()
df_avg_students_per_borough
```

## Summary of grade_span_max grouped by borough

```python
df.groupby("borough")["grade_span_max"].describe()
```

# Visualizations

**Bar chart:** Number of schools per borough

## Number of schools per borough

```python
df_schools_per_borough.plot(kind="bar")

plt.xlabel("Borough")
plt.ylabel("Number of schools")
plt.title("Number of schools per borough")

plt.show()
```

## Avg number of students per borough

```python
df_avg_students_per_borough.plot(kind="bar")

plt.xlabel("Borough")
plt.ylabel("Number of students")
plt.title("Number of students per borough")

plt.show()
```

# Key insights

**1. There is a strong imbalance in school distribution across boroughs.**
  - Brooklyn has the highest number of schools (121), followed closely by the Bronx (118), while Staten Island has by far the fewest schools (10). 

**2. Average student enrollment per school shows an inverse pattern to school counts.**
  - Boroughs with fewer schools tend to have larger schools on average. Staten Island (10 schools) and Queens (80 schools) have the highest average student populations per school, while the Bronx (118 schools) has the lowest.

**3. Grade span structures are highly consistent across boroughs.** 
 - Nearly all schools extend to Grade 12, with only minor variation in lower grade boundaries, indicating a standardized upper-grade offering citywide.
